```python
import pandas as pd
```

```python
cpd = pd.read_csv('cpdata.csv')
ftd = pd.read_csv('Fertilizer.csv')
```

```python
cpd.head()
```

```python
ftd.head()
```

```python
def change_case(i):
    i = i.replace(" ", "")
    i = i.lower()
    return i
```

```python
cpd['label'] = cpd['label'].apply(change_case)
ftd['Crop'] = ftd['Crop'].apply(change_case)
```

```python
ftd['Crop'] = ftd['Crop'].replace('mungbeans','mungbean')
ftd['Crop'] = ftd['Crop'].replace('lentils(masoordal)','lentil')
ftd['Crop'] = ftd['Crop'].replace('pigeonpeas(toordal)','pigeonpeas')
ftd['Crop'] = ftd['Crop'].replace('mothbean(matki)','mothbeans')
ftd['Crop'] = ftd['Crop'].replace('chickpeas(channa)','chickpea')
```

```python
crop_names_from_ftd = ftd['Crop'].unique()
crop_names_from_ftd
```

```python
ftd.tail()
```

```python
for i in crop_names_from_ftd:
    print(cpd[cpd['label'] == i])
```

```python
get_labels = []
for i in crop_names_from_ftd:
    if i in crop_names:
        get_labels.append(i)
```

```python
new_cpd = pd.DataFrame(columns = cpd.columns)
new_ftd = pd.DataFrame(columns = ftd.columns)
```

```python
new_cpd
```

```python
for label in get_labels:
    new_cpd = new_cpd.append(cpd[cpd['label'] == label])
```

```python
for label in get_labels:
    new_ftd = new_ftd.append(ftd[ftd['Crop'] == label].iloc[0])
```

```python
merge_cpd = pd.read_csv('FileCrop.csv')
recom_ftd = merge_ftd
```

```python
import random
temp = pd.DataFrame(columns = ['N','P','K'])
for i in range(0,merge_cpd.shape[0]):
    crop = merge_cpd.label.iloc[i]
    N = recom_ftd[recom_ftd['Crop'] == crop]["N"].iloc[0] + random.randint(-15,20)
    P = recom_ftd[recom_ftd['Crop'] == crop]["P"].iloc[0] + random.randint(-10,15)
    K = recom_ftd[recom_ftd['Crop'] == crop]["K"].iloc[0] + random.randint(-10,5)
    d = {"N":N,"P":P,"K":K}
    temp = temp.append(d,ignore_index = True)
```

```python
temp
```

```python
merge_cpd['N'] = temp['N']
merge_cpd['P'] = temp['P']
merge_cpd['K'] = temp['K']
```

```python
merge_cpd
```

```python
del merge_cpd['Unnamed: 0']
```

```python
merge_cpd
```

```python
merge_cpd = merge_cpd[[ 'N', 'P', 'K','temperature', 'humidity', 'ph', 'rainfall', 'label']]
```

```python
merge_cpd.to_csv("FinalCP.csv",index=False)
```
